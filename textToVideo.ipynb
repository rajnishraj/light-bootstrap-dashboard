```python
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
# from diffusers.utils import export_to_video

pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()


# video_path = export_to_video(video_frames)
# video_path
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

def animate_frames(frames):
    fig = plt.figure()
    ims = []
    for frame in frames:
        im = plt.imshow(frame, animated=True)
        ims.append([im])
    ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True)

    # Save the animation as a video (mp4) file
    video_path = '/content/drive/My Drive/generated_video.mp4'
    ani.save(video_path, writer='ffmpeg')

    # Close the plot to prevent duplicate displays
    plt.close()

    return video_path


prompt = "Narendra modi in pakistan"
video_frames = pipe(prompt, num_inference_steps=25).frames
video_path = animate_frames(video_frames)
video_url = 'https://drive.google.com/uc?export=download&id=' + video_path.split('/')[-1].split('.')[0]
HTML(f'<video src="{video_url}" controls autoplay width="640" height="480"></video>')
```
